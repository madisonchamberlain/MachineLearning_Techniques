# Comparison: The original prediction for reference

```
from sklearn import tree
import time
start_time = time.time()
clf = tree.DecisionTreeClassifier(max_depth=6)

clf = clf.fit(train.iloc[:,0:54], pd.DataFrame(train.iloc[:,54]))
y_pred = clf.predict(train.iloc[:,0:54])
end_time = time.time()
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(pd.DataFrame(train.iloc[:,54]), pd.DataFrame(y_pred))
print("Confusion Matrix")
print(cm)
print("Total correct predictions:")
print(158304+212829+31382+869+567+601+1119)
print("Total time using PCA with 4 components and K = 20")
print(round((end_time - start_time), 4), "seconds")
```

#PCA:

First, I wanted to examenine which and how many of the components were important in generating the forest coverage prediction.  I ran explained_variance_ratio_ on all components to determine how much of the variance was explained by each component.  I found that over 99% of the variance could be explained with the first 4 components.  

```
from sklearn.decomposition import PCA
X = train.iloc[:,0:54]
Y = pd.DataFrame(train.iloc[:,54])
pca = PCA(n_components=54)
pca.fit(X)
X_pca = pca.transform(X)
```

```
print("Explained variance ratio for each component:")
print(pca.explained_variance_ratio_)
print("Explained variance ratio for the first four components combined:")
print(0.66245284 + 0.30733696 + 0.01771567 + 0.00859396)
```

I tried various values of K and found that 20 was the best most often.  This is the confusion matrix, the number correct, and the time it takes to run this method.

```
from sklearn.neighbors import KNeighborsRegressor
import time
start_time = time.time()
knn = KNeighborsRegressor(n_neighbors=20)
knn.fit(X_pca, Y)
pca = PCA(n_components=4)
pca.fit(X)
X_pca = pca.transform(X)
knn.fit(X_pca, Y)
y_pred = knn.predict(X_pca).round()
end_time = time.time()
```

```
from sklearn.metrics import confusion_matrix
print("Confusion matrix:")
print(confusion_matrix(pd.DataFrame(train.iloc[:,54]), pd.DataFrame(y_pred)))
print("Total correct predictions:")
print(184924+261753+26378+1810+2814+5587+1336)
print("Total time using PCA with 4 components and K = 20:")
print(round((end_time - start_time), 4), "seconds")
```

# Random Forest with and without parallel 

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import time


X_train, X_test, y_train, y_test = train_test_split(train.iloc[:,0:54], pd.DataFrame(train.iloc[:,54]), test_size=1000)
start_time = time.time()
clf = RandomForestRegressor()
clf = clf.fit(X_train, y_train.values.ravel())
y_pred = clf.predict(X_test)
end_time = time.time()

from sklearn.metrics import confusion_matrix
print("Confusion matrix (Random Forest):")
print(confusion_matrix(pd.DataFrame(y_test), pd.DataFrame(y_pred.round())))
print("Total time(Random Forest):")
print(round((end_time - start_time), 4), "seconds")


start_time = time.time()
clf = RandomForestRegressor(n_jobs=-1)
clf = clf.fit(X_train, y_train.values.ravel())
y_pred = clf.predict(X_test)
end_time = time.time()
print("Confusion matrix (Random Forest Parallel):")
print(confusion_matrix(pd.DataFrame(y_test), pd.DataFrame(y_pred.round())))
print("Total time(Random Forest Parallel):")
print(round((end_time - start_time), 4), "seconds")
```

# Adaboost:

```
from sklearn.ensemble import AdaBoostClassifier
import time 

X_train, X_test, y_train, y_test = train_test_split(train.iloc[:,0:54], pd.DataFrame(train.iloc[:,54]), test_size=1000)

start_time = time.time()
clf = AdaBoostClassifier(n_estimators=1000)
clf = clf.fit(X_train, y_train.values.ravel())
y_pred = clf.predict(X_test)
end_time = time.time()

print("Confusion matrix (Adaboost):")
print(confusion_matrix(pd.DataFrame(y_test), pd.DataFrame(y_pred.round())))
print("Total time(Adaboost):")
print(round((end_time - start_time), 4), "seconds")
```
